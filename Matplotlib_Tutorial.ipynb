```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

Line Graph

```python
x=np.array([0,6,8,9,4])
```

```python
plt.plot(x)
```

Bar Plot

```python
x=np.array([1,2,3,4])
y=np.array([7,8,9,6])
```

```python
plt.bar(x,y)
```

Scatter Plot

```python
x=np.array([1,2,3,4])
y=np.array([7,8,9,6])
```

```python
plt.scatter(x,y)
```

Pie Chart

```python
x=np.array([1,2,3,4])
plt.pie(x)
```

Histogram

```python
x=np.random.normal(170,10,8)
plt.hist(x)

```
