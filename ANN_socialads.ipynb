```python
import tensorflow as tf
from keras import Sequential
from tensorflow import keras
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(128,activation='relu',input_dim=12))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Import necessary modules
import numpy as np

# Define training data and labels
x_train = np.random.rand(100, 12)  # Replace with actual training data
y_train = np.random.randint(0, 2, size=(100, 1))  # Replace with actual training labels

# Define validation data and labels
x_val = np.random.rand(50, 12)  # Replace with actual validation data
y_val = np.random.randint(0, 2, size=(50, 1))  # Replace with actual validation labels

# Fit the model
history = model.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=100)
```
