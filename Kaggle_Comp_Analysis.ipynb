Logistic Reegression

```python
df = pd.read_csv('/content/WNCAATourneyDetailedResults.csv')
#df.head()  #shape , describe
df = df[['WScore','LScore','WFGM','NumOT']]

#Defining X and Y
X = np.asarray(df[['WScore','LScore','WFGM']])
#X[0:10]
Y = np.asarray(df[['NumOT']])
#Y[0:10]

#Normalizing
X = preprocessing.StandardScaler().fit(X).transform(X)
#X[0:5]
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.1,random_state=4)
print('Train set : ',X_train.shape,Y_train.shape)
print('Test shape : ',X_test.shape,Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Fitting Model with train set
LR = LogisticRegression(C=0.01,solver = 'liblinear').fit(X_train,Y_train)
#predicting
Yhat = LR.predict(X_test)
yhat_prob = LR.predict_proba(X_test)
yhat_prob
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_train,Y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(X_test,Y_test)
```
